```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data representing network performance metrics
# You can replace this with actual data from your research
protocols = ['HTTP', 'FTP', 'SMTP', 'DNS']
average_latency = [100, 150, 120, 80]  # in milliseconds
packet_loss = [0.5, 1.2, 0.8, 0.3]  # in percentage

# Plotting average latency
plt.figure(figsize=(8, 5))
plt.bar(protocols, average_latency, color='skyblue')
plt.title('Average Latency of Application Layer Protocols')
plt.xlabel('Protocols')
plt.ylabel('Average Latency (ms)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plotting packet loss
plt.figure(figsize=(8, 5))
plt.bar(protocols, packet_loss, color='salmon')
plt.title('Packet Loss Rate of Application Layer Protocols')
plt.xlabel('Protocols')
plt.ylabel('Packet Loss (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
